```python
# train_fixed_rotation.py

import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 1) actions 리스트와 time_num들
actions = [
    'ㄱ','ㄴ','ㄷ','ㄹ','ㅁ','ㅂ','ㅅ','ㅇ','ㅈ','ㅊ','ㅋ','ㅌ','ㅍ','ㅎ',
    'ㅏ','ㅑ','ㅓ','ㅕ','ㅗ','ㅛ','ㅜ','ㅠ','ㅡ','ㅣ',
    'ㅐ','ㅒ','ㅔ','ㅖ','ㅢ','ㅚ','ㅟ'
]
time_num, time_num1, time_num2 = '1748496945', '1748529371', '1748533965'
DATA_DIR = "dataset"  # .npy 파일들이 있는 폴더

# 2) 지정된 npy 파일만 불러오는 함수
def load_specified_data():
    arrays = []
    # 첫 번째 time_num
    for act in actions:
        path = os.path.join(DATA_DIR, f"seq_{act}_{time_num}.npy")
        if not os.path.isfile(path):
            raise FileNotFoundError(f"파일이 없습니다: {path}")
        arrays.append(np.load(path))  # shape=(samples_i, 10, 56)

    # 두 번째 time_num
    for act in actions:
        path = os.path.join(DATA_DIR, f"seq_{act}_{time_num1}.npy")
        if not os.path.isfile(path):
            raise FileNotFoundError(f"파일이 없습니다: {path}")
        arrays.append(np.load(path))

    # 세 번째 time_num
    for act in actions:
        path = os.path.join(DATA_DIR, f"seq_{act}_{time_num2}.npy")
        if not os.path.isfile(path):
            raise FileNotFoundError(f"파일이 없습니다: {path}")
        arrays.append(np.load(path))

    # concat → data_all.shape == (30649, 10, 56) 등
    data_all = np.concatenate(arrays, axis=0)
    print(f"합쳐진 데이터 shape (samples, seq_len, 56): {data_all.shape}")

    # x_data: 맨 앞 55개 feature  → (samples, 10, 55)
    x_data = data_all[:, :, :-1].astype(np.float32)

    # y_data: data_all[:, 0, -1]  → (samples,)
    y_data = data_all[:, 0, -1].astype(np.int32)

    print(f"x_data shape: {x_data.shape}, y_data shape: {y_data.shape}")
    return x_data, y_data

# 3) “feature 단계에서 90° 복원” 함수
def rotate_back_features(x_data: np.ndarray) -> np.ndarray:
    """
    x_data: np.ndarray of shape (N, seq_len, 55)
      - [:,:, :40] 은 20쌍(normalized joint-vector)의 (vx, vy)
      - [:,:, 40:55] 은 15개 angle 정보 (degree)

    로그 결과에 따라 “(vx', vy') = (vy, -vx)” 식으로 복원합니다.
    """
    N, T, D = x_data.shape
    assert D == 55, f"feature dimension이 55가 아닙니다: {D}"

    # (N*T, 20, 2) 형태로 reshape
    vectors = x_data[..., :40].reshape(-1, 20, 2)  # shape = (N*T, 20, 2)
    vx = vectors[..., 0]  # shape = (N*T, 20)
    vy = vectors[..., 1]  # shape = (N*T, 20)

    # (vx', vy') = (vy, -vx)
    vx_prime = vy
    vy_prime = -vx

    restored_vectors = np.stack([vx_prime, vy_prime], axis=-1)  # (N*T, 20, 2)
    restored_flat = restored_vectors.reshape(N, T, 40)         # (N, T, 40)

    # angle(마지막 15개) 정보는 그대로
    angles = x_data[..., 40:55]  # shape = (N, T, 15)

    # 복원된 벡터(40) + angle(15) 합쳐서 (55)
    x_fixed = np.concatenate([restored_flat, angles], axis=-1)
    return x_fixed

# 4) F1 Score Metric 정의
def metric_f1score(y_true, y_pred):
    y_pred_rounded = tf.round(y_pred)
    TP = tf.reduce_sum(y_true * y_pred_rounded)
    FP = tf.reduce_sum((1 - y_true) * y_pred_rounded)
    FN = tf.reduce_sum(y_true * (1 - y_pred_rounded))
    precision = TP / (TP + FP + 1e-6)
    recall = TP / (TP + FN + 1e-6)
    F1 = 2 * precision * recall / (precision + recall + 1e-6)
    return F1

# 5) LSTM 모델 정의 (원본 아키텍처 그대로)
def build_lstm_model(input_shape, num_classes):
    model = Sequential([
        LSTM(
            64,
            activation='relu',
            input_shape=input_shape,
            kernel_regularizer=tf.keras.regularizers.l2(0.01)
        ),
        Dropout(0.3),
        Dense(
            32,
            activation='relu',
            kernel_regularizer=tf.keras.regularizers.l2(0.01)
        ),
        Dropout(0.3),
        Dense(num_classes, activation='softmax',
              kernel_regularizer=tf.keras.regularizers.l2(0.01))
    ])
    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy', metric_f1score]
    )
    return model

# 6) 메인 함수: 데이터 로드 → rotate → split → 학습 → 저장 → TFLite 변환
def main():
    # (1) 지정된 데이터 로드
    x_data, y_data = load_specified_data()  # (N, 10, 55), (N,)

    # (2) feature 단계에서 90° 복원 (로그 값에 맞춘 부분)
    x_fixed = rotate_back_features(x_data)  # (N, 10, 55)

    # (3) one-hot 인코딩
    num_classes = len(actions)
    y_onehot = to_categorical(y_data, num_classes)

    # (4) train/val split (레이블 분포 고르게 stratify=y_data)
    x_train, x_val, y_train, y_val = train_test_split(
        x_fixed, y_onehot,
        test_size=0.2,
        random_state=2020,
        stratify=y_data
    )
    print(f"x_train: {x_train.shape}, y_train: {y_train.shape}")
    print(f"x_val:   {x_val.shape},   y_val:   {y_val.shape}")

    # (5) 모델 생성
    input_shape = x_train.shape[1:]  # (10, 55)
    model = build_lstm_model(input_shape, num_classes)
    model.summary()

    # (6) 콜백 설정
    os.makedirs("models", exist_ok=True)
    checkpoint_cb = ModelCheckpoint(
        filepath="models/multi_hand_gesture_classifier_rot_fixed.h5",
        monitor='val_accuracy',
        mode='max',
        save_best_only=True,
        verbose=1
    )
    reduce_lr_cb = ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=3,
        mode='min',
        verbose=1
    )
    earlystop_cb = EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True,
        mode='min',
        verbose=1
    )
    callbacks = [checkpoint_cb, reduce_lr_cb, earlystop_cb]

    # (7) 모델 훈련
    history = model.fit(
        x_train, y_train,
        validation_data=(x_val, y_val),
        epochs=200,
        batch_size=16,
        callbacks=callbacks
    )

    # (8) 학습 결과 시각화 (필요하면 생략 가능)
    plt.figure(figsize=(12, 6))
    plt.plot(history.history['loss'], 'r-', label='Train Loss')
    plt.plot(history.history['val_loss'], 'r--', label='Val Loss')
    plt.plot(history.history['accuracy'], 'b-', label='Train Acc')
    plt.plot(history.history['val_accuracy'], 'b--', label='Val Acc')
    plt.xlabel('Epoch')
    plt.ylabel('Value')
    plt.legend(loc='upper right')
    plt.title('Training & Validation Loss/Acc')
    plt.show()

    # (9) 최종 H5 모델 저장
    model.save("models/multi_hand_gesture_classifier_rot_fixed.h5")
    print("✅ 최종 H5 모델 저장: models/multi_hand_gesture_classifier_rot_fixed.h5")

    # (10) TFLite 변환 (TensorList ops 지원 활성화 포함)
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    converter.experimental_enable_resource_variables = True
    converter.target_spec.supported_ops = [
        tf.lite.OpsSet.TFLITE_BUILTINS,
        tf.lite.OpsSet.SELECT_TF_OPS
    ]
    converter._experimental_lower_tensor_list_ops = False

    tflite_model = converter.convert()
    with open("models/multi_hand_gesture_classifier_rot_fixed.tflite", "wb") as f:
        f.write(tflite_model)
    print("✅ TFLite 모델 저장: models/multi_hand_gesture_classifier_rot_fixed.tflite")


if __name__ == "__main__":
    main()
```
